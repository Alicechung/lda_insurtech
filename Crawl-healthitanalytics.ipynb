```python
healthitlinkls = []
for j in range(1,114):
    link = "https://healthitanalytics.com/news/P%s/" % (20*j)
    healthitlinkls.append(link)
```

```python
healthitlinkls[61]
```

```python
browser.find_element_by_xpath('//*[@id="news-articles"]/div[3]/div/div[1]/h3/a')
```

```python
browser.find_element_by_xpath()
```

```python
whole_art_num = 1238
#os.mkdir("%s"%"healthit")
#os.chdir("%s"%"healthit")
for page_num in range(61, len(healthitlinkls)+1):
    if page_num == 0:
        pass
    else:
        browser.get(healthitlinkls[page_num-1])
        #print('next',page_num+2)
        time.sleep(1)

    articlels = browser.find_elements_by_tag_name('h3')    
    article_num = len(articlels)
    
    for k in range(1, article_num+1):
        whole_art_num += 1
        if page_num == 5 and k == 3:
            article = browser.find_element_by_xpath('//*[@id="news-articles"]/div[3]/div/div[1]/h3/a')
        else:
            article = browser.find_element_by_xpath('//*[@id="news-articles"]/div[%s]/div/h3/a'%k)        
        article.click()
        
        #browser.get(article)    
        filename = "healthit_article_"+str(whole_art_num)+'.txt'
        with open(filename, 'w', encoding='UTF-8') as txt_file:
            try:
                header = browser.find_element_by_tag_name('h1').text
                #author = browser.find_element_by_class_name('author').text
                pub_date = browser.find_element_by_tag_name('time').text 
                textbody = browser.find_element_by_tag_name('main').text
                print(whole_art_num, header, pub_date)

                txt_file.write("%s\n" % header)
                txt_file.write("\n")
                #txt_file.write("%s\n" % author)
                #txt_file.write("\n") 
                txt_file.write("%s\n" % pub_date)
                txt_file.write("\n") 
                txt_file.write("%s\n" % textbody)
                txt_file.write("\n\n")
                time.sleep(1)
            except Exception as e:
                print(page_num, article_num)
                print(whole_art_num)
                print(e)

            txt_file.close()
        browser.back()
        time.sleep(1)
```
